# OUTLIERS REMOVAL

```python
def remove_outliers(dfout, feature_names):
    df_no_outliers = df11.copy()


    for feature_name in feature_names:
        Q1 = df_no_outliers[feature_name].quantile(0.25)
        Q3 = df_no_outliers[feature_name].quantile(0.75)
        IQR = Q3 - Q1

        # Remove outliers
        df_no_outliers = df_no_outliers[(df_no_outliers[feature_name] >= Q1 - 1.5 * IQR) & (df_no_outliers[feature_name] <= Q3 + 1.5 * IQR)]


    return  df_no_outliers 


features_to_process = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year']


# Remove outliers for all specified features
df_no_outliers  = remove_outliers(df11, features_to_process)
```

```python
numerical_columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year']
plt.figure(figsize=(12, 8))
for column in numerical_columns:
    sns.boxplot(x=df_no_outliers[column])
    plt.title(f'Box Plot for {column}')
    plt.show()
```
